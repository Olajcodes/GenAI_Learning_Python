```python
# Setup
import os
from dotenv import load_dotenv

load_dotenv()

openai_api_key = os.getenv("OPENAI_API_KEY")

if not openai_api_key:
    print("Warning: OPENAI_API_KEY not found. Set it in .env file.")
else:
    print("API Key loaded successfully!")
```

```python
# Sample documents for all examples
documents = [
    "Python is a high-level programming language. I so much love it",
    "Learning Python the very hard way!!",
    "JavaScript can be so funny and interesting same time."
    ]
```

**Approach 1: Custom Implementation (From Module 6)**

```python
# Custom RAG
from sentence_transformers import SentenceTransformer
import numpy as np
from openai import OpenAI

# Generate embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
doc_embeddings = model.encode(documents)
# print(doc_embeddings)
```

```python
# Query and retrieve
query = "What is Python?"
query_embedding = model.encode([query])[0]
# print(query_embedding)
```

```python
# Compare Similarity
similarities = np.dot(doc_embeddings, query_embedding)
top_idx = np.argmax(similarities)  #Returns the indices of the top most similar doc
retrieved_doc = documents[top_idx]  #Stores the retrieved document using it's index
```

```python
retrieved_doc
```

```python
# Generate response
client = OpenAI(api_key=openai_api_key)
prompt = f"""Context: {retrieved_doc}
Question: {query}
Answer based on the context"""

response = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=[{"role": "user", "content": prompt}],
    temperature=0
)
```
